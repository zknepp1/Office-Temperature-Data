```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The purpose of this analysis is to collect temperature data within the office to show how hot this office gets on a daily basis. SOS!!!

Previous research shows that optimal working temperatures for workers in the office are between 70ºF and 72ºF. ***"Any temperature above 73ºF decreases your productivity."*** - https://chrisbailey.com/temperature-productivity/

Below is that data collected from the office. T1, T2, T3, T4 are time variables and represent the time the temperature was recorded. T1 and T2 are temperature recorded in the morning. T3 and T4 are temperatures recorded in the afternoon. All temperatures are recorded in Fahrenheit. The weather variable shows whether the sun was beaming into the office in the afternoon (s) or if it was cloudy (c). It is found that the weather was the biggest determination of whether the temperature was going to reach above 80 ºF.

The data collected from the office show the temperature nearly every every day reaching above 73ºF.

```python
# Data collected from the office
T1 = [72.4, 69.4, 69.7, 69.1, 73.5, 72.4, 72.7, 71.6, 71.0, 71.6, 71.0, 70.8, 70.2, 70.3, 71.0, 72.4, 71.5, 72.0, 73.4, 73.5, 72.1, 71.9]
T2 = [74.6, 72.2, 73.0, 72.2, 75.3, 76.4, 75.1, 71.7, 72.9, 73.3, 72.8, 71.8, 72.2, 72.5, 73.2, 73.9, 73.0, 73.7, 74.7, 75.3, 73.9, 74.3]
T3 = [76.6, 76.9, 75.0, 72.5, 76.5, 79.9, 82.1, 75.3, 78.2, 75.7, 73.1, 72.4, 77.1, 78.6, 80.5, 78.3, 80.7, 79.4, 79.2, 75.5, 74.6, 74.5]
T4 = [79.6, 80.1, 75.5, 73.2, 76.7, 82.8, 83.0, 79.5, 78.8, 79.3, 73.3, 72.2, 79.0, 81.1, 82.7, 79.8, 82.0, 80.6, 82.7, 75.2, 75.3, 78.6]
weather = ['s','s','c','c','c','s','s','s','s','s','c','c','s','s','s','s','s','s','s','c','c','c']
```

```python
# Double checking the lengths of the data. They should all be the same.
print(len(T1))
print(len(T2))
print(len(T3))
print(len(T4))
print(len(weather))
```

```python
# Turning data into a dataframe, and double checking.
df = pd.DataFrame({'T1':T1,'T2':T2,'T3':T3,'T4':T4,'weather':weather})
print(df.shape)
df.head()
```

```python
#Line Plot of the different time groups
plt.figure(figsize=(10, 6))
for column in df.columns[:-1]:
    plt.plot(df[column], label=column)
plt.xlabel('Observation')
plt.ylabel('Temperature in ºF')
plt.title('Temperature Trends by Time Group')
plt.legend()
plt.show()
```

The above plot shows a line graph of the different time groups. Each observation represents 1 day of data. From the graph, you can observe that the morning time groups (T1 and T2) remain significantly below 78ºF. You can also derive from the graph that the afternoon time groups are consistently hotter than the morning time groups often reaching above 80ºF. Just why...?

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([df['T1'], df['T2'], df['T3'], df['T4']], labels=['T1', 'T2', 'T3', 'T4'])
plt.title('Temperature Distribution by Time Group')
plt.xlabel('Time Group')
plt.ylabel('Temperature in ºF')
plt.show()
```

The above plot shows a boxplot comparing the different time groups. This gives a different visual of how the time groups are distrubuted. The morning groups T1 and T2 show a significant lower tempurature than the afternoon groups. Over the cours of the 4 time groups, we can clearly see a pattern that throughout the day, the temperatures on average will increase.

```python
# Separate T4 temperatures based on weather conditions
t4_sunny = df[df['weather'] == 's']['T4']
t4_cloudy = df[df['weather'] == 'c']['T4']

# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([t4_sunny, t4_cloudy], labels=['Sunny', 'Cloudy'])
plt.title('Temperature Distribution in T4 for Different Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Temperature in ºF')
plt.show()
```

The final plot shows a boxplot of the last time group (T4) recorded for that day compared to the weather recorded for that day. From the graph you can observe that the sunny days made the office significantly hotter  than the cloudy days with the temperature often reaching above a high temperature of 80ºF, and sometimes getting above 82ºF.
